```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Tahun': [2021, 2022, 2023, 2024, 2025],
    'Penjualan': [100, 130, 150, , 300]
}

df = pd.DataFrame(data)

df.to_csv('penjualan.csv', index=False)
print("File CSV berhasil dibuat: penjualan.csv")

plt.plot(df['Tahun'], df['Penjualan'], marker='o', color='green')
plt.title('Data Penjualan')
plt.xlabel('Tahun')
plt.ylabel('Penjualan')
plt.grid(True)

plt.show()
```
